````python
# Извлекаем результаты SQL-запроса
query = '''
SELECT 
    p.user_id,
    p.device_type_canonical,
    p.order_id,
    p.created_dt_msk as order_dt,
    p.created_ts_msk as order_ts,
    p.currency_code,
    p.revenue,
    p.tickets_count,
    -- Добавленное поле: количество дней с предыдущей покупки
    CASE 
        WHEN LAG(p.created_dt_msk) OVER (PARTITION BY p.user_id ORDER BY p.created_dt_msk) IS NOT NULL
        THEN EXTRACT(DAY FROM p.created_dt_msk - LAG(p.created_dt_msk) OVER (PARTITION BY p.user_id ORDER BY p.created_dt_msk))
        ELSE NULL
    END::INTEGER as days_since_prev,
    p.event_id,
    e.event_name_code as event_name,
    e.event_type_main,
    p.service_name,
    r.region_name,
    c.city_name
FROM afisha.purchases p
INNER JOIN afisha.events e ON p.event_id = e.event_id
INNER JOIN afisha.city c ON e.city_id = c.city_id
INNER JOIN afisha.regions r ON c.region_id = r.region_id
WHERE 
    -- Фильтрация по типам устройств: только мобильные и десктопные
    p.device_type_canonical IN ('mobile', 'desktop')
    -- Исключаем заказы на фильмы
    AND e.event_type_main != 'фильм'
ORDER BY p.user_id
''' 
````

````python
# Выгружаем данные из датасета с конверсией валюты
tenge_df = pd.read_csv('https://code.s3.yandex.net/datasets/final_tickets_tenge_df.csv')
````

---
Необходимо преобразовать дату в корректный формат
Столбец `data` сейчас текст (object), нужно сделать датой

---

````python
# Gреобразовываем дату в правильный формат

tenge_df['data'] = pd.to_datetime(tenge_df['data'])

# Проверка изменений
print("Тип данных в столбце 'data' после преобразования:")
print(type(tenge_df['data'][0]))
print("\nПервые 5 дат:")
print(tenge_df['data'].head())
````

````python
# Выведем данные по валюте в основном датасете

print("\nТипы валют в основном датасете:")
print(df['currency_code'].value_counts())

# Посмотрим на несколько строк с разными валютами
print("\nПримеры заказов в разных валютах:")
rub_orders = df[df['currency_code'] == 'rub'].head(2)
kzt_orders = df[df['currency_code'] == 'kzt'].head(2)

print("Рубли:")
print(rub_orders[['order_dt', 'revenue', 'currency_code']])

print("\nТенге:")
print(kzt_orders[['order_dt', 'revenue', 'currency_code']])
````

````python
# Создадим словарь, где ключ - дата, значение - курс
kurs_po_date = {}
for index, row in tenge_df.iterrows():
    data = row['data']
    kurs = row['curs']
    kurs_po_date[data] = kurs

print(f"Всего курсов в словаре: {len(kurs_po_date)}")
print("Пример: курс на 2024-01-10:", kurs_po_date.get(pd.to_datetime('2024-01-10')))
````

````python
# Создадим функцию для конвертации
def konvertiruj_v_rubli(stroka):
    valuta = stroka['currency_code']
    vyruchka = stroka['revenue']
    data_zakaza = stroka['order_dt']
    
    if valuta == 'rub':
        # Если уже в рублях, возвращаем как есть
        return vyruchka
    elif valuta == 'kzt':
        # Нужно найти курс на дату заказа
        data_zakaza = stroka['order_dt']
        kurs = kurs_po_date[data_zakaza]  # Берем курс прямо из словаря
        
        # Конвертируем: (выручка в тенге / 100) * курс
        vyruchka_v_rublyah = (vyruchka / 100) * kurs
        return vyruchka_v_rublyah
    
# Применяем функцию
df['revenue_rub'] = df.apply(konvertiruj_v_rubli, axis=1)    
````

### 3. Создание профиля пользователя <a id='3'></a>

В будущем отдел маркетинга планирует создать модель для прогнозирования возврата пользователей. Поэтому сейчас они просят вас построить агрегированные признаки, описывающие поведение и профиль каждого пользователя.

---

**Задача 3.1.** Постройте профиль пользователя — для каждого пользователя найдите:

- дату первого и последнего заказа;
- устройство, с которого был сделан первый заказ;
- регион, в котором был сделан первый заказ;
- билетного партнёра, к которому обращались при первом заказе;
- жанр первого посещённого мероприятия (используйте поле `event_type_main`);
- общее количество заказов;
- средняя выручка с одного заказа в рублях;
- среднее количество билетов в заказе;
- среднее время между заказами.

После этого добавьте два бинарных признака:

- `is_two` — совершил ли пользователь 2 и более заказа;
- `is_five` — совершил ли пользователь 5 и более заказов.

**Рекомендация:** перед тем как строить профиль, отсортируйте данные по времени совершения заказа.

---


````python
# Cортируем данные по времени совершения заказа

df = df.sort_values(by=['user_id', 'order_ts']).reset_index(drop=True)
print(f"Данные отсортированы. Первые 3 строки:")
print(df[['user_id', 'order_ts', 'order_id']].head(3))
````

````python
# Создаем профиль пользователя
user_profiles = []

# Группируем по пользователям
for user_id, group in df.groupby('user_id'):
    # Базовые признаки первого заказа
    first_order = group.iloc[0]
    
    profile = {
        'user_id': user_id,
        'first_order_date': first_order['order_dt'],
        'last_order_date': group.iloc[-1]['order_dt'],
        'first_device': first_order['device_type_canonical'],
        'first_region': first_order['region_name'],
        'first_partner': first_order['service_name'],
        'first_event_type': first_order['event_type_main'],
        
        # Количественные признаки
        'total_orders': len(group),
        'avg_revenue_rub': group['revenue_rub'].mean(),
        'avg_tickets_count': group['tickets_count'].mean(),
    }
    
    # Среднее время между заказами (только для пользователей с >1 заказом)
    if len(group) > 1:
        # Используем days_since_prev, исключая NaN (первый заказ)
        days_between = group['days_since_prev'].dropna().mean()
        profile['avg_days_between_orders'] = days_between
    else:
        profile['avg_days_between_orders'] = None
    
    user_profiles.append(profile)

# Преобразуем в DataFrame
user_profile_df = pd.DataFrame(user_profiles)

print(f"Создано профилей для {len(user_profile_df)} пользователей")
print("\nПервые 3 профиля:")
print(user_profile_df.head(3))
````

__Комментарий от ревьюера №1__

````python
# Добавление бинарных признаков


# is_two — совершил ли пользователь 2 и более заказа
user_profile_df['is_two'] = (user_profile_df['total_orders'] >= 2).astype(int)

# is_five — совершил ли пользователь 5 и более заказов
user_profile_df['is_five'] = (user_profile_df['total_orders'] >= 5).astype(int)

print(f"\nСтатистика по бинарным признакам:")
print(f"Пользователей с 2+ заказами (is_two=1): {user_profile_df['is_two'].sum()} ({user_profile_df['is_two'].mean()*100:.1f}%)")
print(f"Пользователей с 5+ заказами (is_five=1): {user_profile_df['is_five'].sum()} ({user_profile_df['is_five'].mean()*100:.1f}%)")
````

````python
print("ЗАДАЧА 4.3.1: ВЛИЯНИЕ ДНЯ НЕДЕЛИ ПЕРВОЙ ПОКУПКИ НА ВОЗВРАТ")
print("="*75)

# 1. Выделяем день недели из даты первого заказа
user_profiles['first_order_dayofweek'] = user_profiles['first_order_date'].dt.dayofweek
user_profiles['first_order_dayname'] = user_profiles['first_order_date'].dt.day_name()

# Словарь для русских названий дней
days_ru = {
    'Monday': 'Понедельник',
    'Tuesday': 'Вторник',
    'Wednesday': 'Среда',
    'Thursday': 'Четверг',
    'Friday': 'Пятница',
    'Saturday': 'Суббота',
    'Sunday': 'Воскресенье'
}

user_profiles['first_order_dayname_ru'] = user_profiles['first_order_dayname'].map(days_ru)

# 2. Считаем статистику по дням недели
print("\n1. СТАТИСТИКА ПО ДНЯМ НЕДЕЛИ:")
print("="*75)

# Порядок дней для сортировки
day_order = ['Понедельник', 'Вторник', 'Среда', 'Четверг', 'Пятница', 'Суббота', 'Воскресенье']

weekday_stats = []

print(f"{'День недели':<15} {'Пользователей':>15} {'Доля от всех':>15} {'Вернулись':>12} {'Доля %':>12}")
print("-" * 75)

for day in day_order:
    day_users = user_profiles[user_profiles['first_order_dayname_ru'] == day]
    total = len(day_users)
    percent_of_total = (total / len(user_profiles)) * 100
    returns = day_users['is_two'].sum()
    return_rate = (returns / total) * 100
    
    weekday_stats.append([day, total, percent_of_total, returns, return_rate])
    
    print(f"{day:<15} {total:>13,} {percent_of_total:>14.1f}% {returns:>10,} {return_rate:>11.1f}%")

# Общее среднее
total_return_rate = user_profiles['is_two'].mean() * 100
print("-" * 75)
print(f"{'ВСЕГО':<15} {len(user_profiles):>13,} {'':>14} {user_profiles['is_two'].sum():>10,} {total_return_rate:>11.1f}%")
````

````python
# 4. АНАЛИЗ ПО СЕГМЕНТАМ (БЕЗ ИСКУССТВЕННЫХ ЗНАЧЕНИЙ)
print("\n" + "="*70)
print("АНАЛИЗ ПО СЕГМЕНТАМ:")
print("="*70)

segments = {
    '1 заказ': user_profiles[user_profiles['total_orders'] == 1],
    '2-4 заказа': user_profiles[(user_profiles['total_orders'] >= 2) & (user_profiles['total_orders'] <= 4)],
    '5+ заказов': user_profiles[user_profiles['total_orders'] >= 5]
}

for name, data in segments.items():
    print(f"\n{name.upper()} ({len(data):,})")
    print("-" * 40)
    
    # Для сегмента "1 заказ" нет avg_days_between_orders
    if name == '1 заказ':
        data_seg = data[['avg_revenue_rub', 'avg_tickets_count', 
                         'first_device', 'first_event_type', 'first_region', 'first_partner']]
        interval_cols = ['avg_revenue_rub', 'avg_tickets_count']
    else:
        data_seg = data[['avg_revenue_rub', 'avg_tickets_count', 'avg_days_between_orders',
                         'first_device', 'first_event_type', 'first_region', 'first_partner']]
        interval_cols = ['avg_revenue_rub', 'avg_tickets_count', 'avg_days_between_orders']
    
    # Рассчитываем корреляции
    seg_corr = data_seg.phik_matrix(interval_cols=interval_cols)
    
    # Визуализация
    plt.figure(figsize=(8, 6))
    plot_correlation_matrix(seg_corr.values,
                           x_labels=seg_corr.columns,
                           y_labels=seg_corr.index,
                           vmin=0, vmax=0.5,
                           color_map='Blues',
                           title=f'Корреляция признаков: {name}')
    plt.tight_layout()
    plt.show()
    
    # Показываем сильные корреляции (>0.3)
    print("  Сильные корреляции (>0.3):")
    corr_found = False
    
    for i in range(len(seg_corr.columns)):
        for j in range(i+1, len(seg_corr.columns)):
            corr_val = seg_corr.iloc[i, j]
            if corr_val > 0.3:
                print(f"    {seg_corr.columns[i]} — {seg_corr.columns[j]}: {corr_val:.3f}")
                corr_found = True
    
    if not corr_found:
        print("    Сильных корреляций не обнаружено")
````
